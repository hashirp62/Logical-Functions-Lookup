5. company has set the following commission structure for its sales team:

 > The commission for the salesperson achieved **45000** will be **10%**

 to find this using **VLOOKUP** use the formula:

  **=VLOOKUP(45000,A2:B5,2,TRUE)**

  45000= lookup value

  A2:B5= lookup array

  2= table to look the commission percentage

  TRUE= 45000 is not exactly given in the table. So TRUE is used to pull the approximate match

 A2:A5 is the the row showing the EmpID to search

 B2:B5 is the row where excel searches the department that is the value to return
